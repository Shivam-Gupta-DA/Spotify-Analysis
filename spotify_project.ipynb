```python
tracks.info()
```

```python
for i in ['popularity','tempo']:
    q1 = tracks[i].quantile(0.25)
    q3 = tracks[i].quantile(0.75)
    
    iqr = q3-q1

    lb=q1-iqr*1.5
    ub=q3+iqr*1.5


    tracks[i] = np.where(tracks[i]<lb, lb,np.where(tracks[i]>ub,ub,tracks[i]))
```

```python
for i in ['popularity','tempo']:
    plt.figure(figsize=(10,6))
    sns.boxplot(tracks[i])
```

#### Explicit tracks have higher energy than non-explicit tracks.

```python
from scipy.stats import ttest_ind

# Separate explicit and non-explicit tracks
explicit_tracks = df[df['explicit'] == 1]['energy']
non_explicit_tracks = df[df['explicit'] == 0]['energy']

# Perform t-test to compare the means
t_stat, p_value = ttest_ind(explicit_tracks, non_explicit_tracks, equal_var=False)

# Output the results
t_stat, p_value
print('p value :',p_value)
print('t test :',t_stat)

if p_value<0.05:
    print('Explicit tracks truly have higher energy.')
```
